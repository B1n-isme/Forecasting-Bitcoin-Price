```python
import pandas as pd

# Read the CSV file
df1 = pd.read_csv("lstm_metrics.csv", index_col="Model_Type")  # Replace with your actual file path
df2 = pd.read_csv("tft_metrics.csv", index_col="Model_Type")  # Replace with your actual file path

# Drop "Best params" column
df1 = df1.drop(columns=["Best params"])
df2 = df2.drop(columns=["Length"])

# # Set "Model_Type" as the index
# df.set_index("Model_Type", inplace=True)

# Rename index as 'Metric'
df1.index.name = "Metric"
df2.index.name = "Metric"

# Transpose the DataFrame
df1 = df1.T
df2 = df2.T

# Concatenate the DataFrames
df = pd.concat([df1, df2], axis=1)

df
```
