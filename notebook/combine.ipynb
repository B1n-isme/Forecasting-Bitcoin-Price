```python
import pandas as pd
from functools import reduce

# List of CSV files to join
csv_files = [
    "bitcoin_daily_google_trends.csv", #
    "blockchain_metrics_updated.csv", #
    "btc_usd_with_indicators_updated.csv", #
    "gold_oil_prices.csv", #
    "macro_financial_factors.csv" #
]

# Load each CSV into a DataFrame and ensure 'Date' is the index
dataframes = []
for file in csv_files:
    df = pd.read_csv(f"../data/processed/{file}", parse_dates=['Date'], index_col='Date')
    dataframes.append(df)

# Merge all DataFrames on 'Date'
merged_df = reduce(lambda left, right: left.join(right, how='outer'), dataframes)

# Save the merged DataFrame to a new CSV
merged_df.to_csv("dataset.csv")

# Display merged DataFrame info
print(merged_df.info())
print(merged_df.head())
```

```python
import pandas as pd

# Read the merged CSV file
df = pd.read_csv("dataset.csv")

# Move 'btc_close' to the end
if 'btc_close' in df.columns:
    btc_close = df.pop('btc_close')  # Remove 'btc_close' column and store it
    df['btc_close'] = btc_close      # Add it back at the end
    print("'btc_close' moved to the end.")
else:
    print("'btc_close' column not found in the file.")

# Save the updated DataFrame back to a CSV
df.to_csv("../data/final/dataset.csv", index=False)

# Display the updated DataFrame info
print(df.head())
```
